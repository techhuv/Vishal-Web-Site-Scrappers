```python
from selenium import webdriver
import pandas as pd
```

```python
import requests
headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/12.1.1 Safari/605.1.15","Accept-Language": "en-gb","Accept-Encoding":"br, gzip, deflate","Accept":"test/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8","Referer":"http://www.google.com/"}
url = "https://www1.nseindia.com/products/dynaContent/equities/equities/bulkdeals.jsp?symbol=&segmentLink=13&symbolCount=&dateRange=12month&fromDate=&toDate=&dataType=DEALS"
page = requests.get(url,headers=headers)
page
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Initailizing the storage
date = []
symbol = []
security_name = []
client_name = []
buy_sell = []
qty_traded= []
trade_price = []
remarks = []
```

```python
tr_elements = soup.findAll('tr')
```

```python
df=pd.DataFrame()
```

```python
tr_elements[3:]
```

```python
for container in tr_elements[3:]:
    #date
    date.append(container.find(class_='date').text)
    
    #symbol, security_name, client_name
    names = container.findAll(class_='normalText')
    symbol.append(names[0].text)
    security_name.append(names[1].text)
    client_name.append(names[2].text)
    
    #buy_sell, qty_traded, traded_price
    numbers = container.findAll(class_='number')
    buy_sell.append(numbers[0].text)
    qty_traded.append(numbers[1].text)
    trade_price.append(numbers[2].text)
    
    #remarks
    remarks.append(names[3].text)
```

```python
'''
print(len(date))
print(len(symbol))
print(len(security_name))
print(len(client_name))
print(len(buy_sell))
print(len(qty_traded))
print(len(trade_price))
print(len(remarks))
'''
```

```python
df['Date']=date
df['Symbol']=symbol
df['Security Name']=security_name
df['Client Name']=client_name
df['Buy/Sell']=buy_sell
df['Qty Traded']=qty_traded
df['Trade Price']=trade_price
df['Remarks']=remarks
```

```python
#Data Cleaning
df['Qty Traded'] = df['Qty Traded'].str.replace(',','').astype(int)
df['Trade Price'] = df['Trade Price'].astype(float)
```
