```python
from selenium import webdriver
import pandas as pd
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup, NavigableString, Tag 
```

```python
with open('D:\protonsHub\Just Eat\justeat.html', 'r') as f:
    contents = f.read()
    soup = BeautifulSoup(contents, 'lxml')
    menu = soup.find('div',attrs = {'data-test-id':'menu-tab'})
```

```python
item_name = submenu[1].find('h4',attrs = {'class':'c-menuItems-heading u-spacingRight'}).text
item_name = re.sub('\s+',' ',item_name).lstrip().rstrip()
item_price = submenu[1].find('p',attrs={'class':'c-menuItems-price notranslate'}).text
item_price = re.sub('\s+',' ',item_price).lstrip().rstrip()
item_desc = submenu[1].find('p',attrs={'class':'u-spacingTop'}).text
item_desc = re.sub('\s+',' ',item_desc).lstrip().rstrip()
```

```python
for r in submenu:
    c= r.header.h3.text
    c= re.sub('\s+',' ',c).lstrip().rstrip()
    items = r.findAll('div',attrs = {'data-test-id':'menu-item'})
    for item in items:
        cat.append(c)
        
        item_name = item.find('h4',attrs = {'class':'c-menuItems-heading u-spacingRight'})
        item_name = item_name.text
        item_name = re.sub('\s+',' ',item_name).lstrip().rstrip()
        itemName.append(item_name)
        
        item_price = item.find('p',attrs={'class':'c-menuItems-price notranslate'})
        item_price = item_price.text
        item_price = re.sub('\s+',' ',item_price).lstrip().rstrip()
        price.append(item_price)
        
        item_desc = item.find('p',attrs={'class':'u-spacingTop'})
        if item_desc is not None:
            item_desc = item_desc.text
            item_desc = re.sub('\s+',' ',item_desc).lstrip().rstrip()
            desc.append(item_desc)
        else:
            desc.append(' - ')
        
    
#         item_name = re.sub('\s+',' ',item_name).lstrip().rstrip()
#         itemName.append(item_name)
        
```

```python
df=pd.DataFrame()
df['category']=cat
df['item name']=itemName
df['price in pounds']=price
df['desc']=desc
```
